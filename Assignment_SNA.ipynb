## 1. Graph Generators를 사용한 그래프 특성 분석

- 18가지 그래프 특성 분석
    - http://networkx.lanl.gov/reference/generators.html
    - Classic으로 분류된 그래프 중 5개 선택
    - Random으로 분류된 그래프 중 5개 선택
    - 총 10개의 그래프를 Generation 하여 각 그래프에 대한 다음을 정리
        - 그래프 모습 캡쳐
        - radius 값
        - diameter 값
        - center 노드
        - periphery 노드
        - 각 노드들에 대한 Clustering Coefficient
        - 그래프에 대한 Average Clustering Coefficient
    - 반드시 각 그래프들에 대한 위 값들을 비교해보고 나름대로 고찰한 분석 결과를 작성하여 넣을 것

### Classic

```python
import networkx as nx
import matplotlib.pyplot as plt
import networkx.algorithms as algo
%matplotlib inline
```

```python
def result(g):
    print "radius(g):", algo.radius(g)
    print "diameter(g):", algo.diameter(g)
    print "center(g):", algo.center(g)
    print "periphery(g):", algo.periphery(g)
    print "clustering(g):",nx.clustering(g)
    print "Average Clustering Coefficient(g):",sum(nx.clustering(g).values())/len(nx.clustering(g))
```

```python
g= nx.barbell_graph(5,2)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 두번째 파라미터의 값만큼의 노드의수로 구성된 직선그래프를 중심으로 양쪽에 첫번째 파라미터의 값만큼의 노드수로 구성된 완전그래프로 이루어져있다. 양쪽에 두개의 완전그래프가 있으므로  Average Clustering Coefficient값은 높은값이 나온다.

```python
g = nx.complete_graph(8)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 완전그래프를 그리는 함수이다.
그래서 모든 수치 값이 높게 나올수 밖에없다.

```python
g =nx.complete_bipartite_graph(2,8)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 첫번째 파라미터값의 수만큼 노드가 중심에 있고 두번째 파라미터 값이 주위를 감싸는 성형 구조이다.
따라서 Diameter는  2가 나올수밖에없다. 

```python
g = nx.circular_ladder_graph(5)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 파라미터 값만큼 정다각형기둥을 만드는 그래프이다.
따라서 Average Clustering Coefficient값이 0이 나올수밖에없다.

```python
g = nx.cycle_graph(10)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 파라미터 값만큼의 노드로 구성된 단순 원형그래프이다.
그래서 diameter의 값은 파라미터의 절반이되고, Average Clustering Coefficient는 0이된다.

### Random

```python
g = nx.gnp_random_graph(10,5)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 radius :1, diameter:1, Average Clustering Coefficient:1 이므로 완전그래프이다.

```python
g = nx.binomial_graph(10,5)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 radius :1, diameter:1, Average Clustering Coefficient:1 이므로 완전그래프이다.

```python
g = nx.dense_gnm_random_graph(5,8)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 첫번째파라미터값만큼 노드를 구성하고 두번째 파라미터 값만은 간선을 구성한다.
위의 그래프는 파라미터값을 완전그래프와 비슷하게 값을 주었으므로  Average Clustering Coefficient 값이 높게 나왔다.

```python
g = nx.gnm_random_graph(7,10)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 첫번째파라미터값만큼 노드를 구성하고 두번째 파라미터 값만은 간선을 구성한다.
그러나 바로 전의 그래프와는 다르게 밀도가 낮은 것을 확인할수 있다.
따라서 Average Clustering Coefficient 값 또한 바로 전의 그래프보다 낮다.

```python
g = nx.erdos_renyi_graph(10,5)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

위의 그래프는 데칼코마니와 같은 특징을 띄고 있는 완전그래프이다.
(radius :1, diameter:1, Average Clustering Coefficient:1  이므로)


## 2. Twitter 소셜 그래프 특성 분석 (1/3)

- 활용할 데이터 집합
    -  http://snap.stanford.edu/data/egonets-Twitter.html
- SCC (Strongly Connected Components)
    - A directed graph is “strongly connected” if it contains a directed path from u to v and a directed path from v to u for every pair of vertices u and v. 
    - The strongly connected components are the maximal strongly connected subgraphs.
- WCC (Weakly Connected Components)
    - A directed graph is “weakly connected” if replacing all of its directed edges with undirected edges produces a connected graph.
    - The weakly connected components are the maximal weakly connected subgraphs.

```python
import snap
```

```python
Graph = snap.LoadConnList(snap.PNGraph, "twitter_combined.txt")
```

```python
type(Graph)
```

```python
#노드의 수
Graph.GetNodes()
```

```python
#간선의 수
Graph.GetEdges()
```

```python
#Average clustering coefficient
snap.GetClustCf(Graph)
```

```python
#Radius 
rad =float("inf")
Graph.Nodes()
for NI in Graph.Nodes():
    GetNodeEcc= snap.GetNodeEcc(Graph, NI.GetId(), True)
    if GetNodeEcc>0 and GetNodeEcc <rad:
        rad = GetNodeEcc
        #print rad

print 'Radius : '+ str(rad)
```

```python
#Diameter
#Returns a 90th percentile of the shortest path length distribution of a Graph
snap.GetAnfEffDiam(Graph)
```

```python
#Nodes in largest SCC
snap.GetMxScc(Graph).GetNodes()
```

```python
#Edges in largest SCC
snap.GetMxScc(Graph).GetEdges()
```

```python
#Nodes in largest WCC
snap.GetMxWcc(Graph).GetNodes()
```

```python
#Edges in largest WCC
snap.GetMxWcc(Graph).GetEdges()
```

```python
Graph = snap.LoadConnList(snap.PUNGraph, "twitter_combined.txt")
```

```python
#Top 10 Nodes of Degree Centrality
dic ={}
for NI in Graph.Nodes():
    DegCentr = snap.GetDegreeCentr(Graph, NI.GetId())
    dic[NI.GetId()]=DegCentr
arr =dic.values()
arr.sort(reverse=True)
for i in range(10):
    print arr[i]
```

```python
#Top 10 Nodes of Closeness Centrality

dic ={}
for NI in Graph.Nodes():
    CloseCentr = snap.GetClosenessCentr(Graph, NI.GetId())
    dic[NI.GetId()]=CloseCentr
arr =dic.values()
arr.sort(reverse=True)
for i in range(10):
    print arr[i]
```

```python
#Top 10 Nodes of Betweenness Centrality

i=0
Nodes = snap.TIntFltH()
Edges = snap.TIntPrFltH()
snap.GetBetweennessCentr(Graph, Nodes, Edges, 1.0)
Nodes.SortByDat(False)
for node in Nodes:
    print "node: %d ,BetweennessCentrality: %f" % (node, Nodes[node])
    if(i>=9):
        break
    else:
        i+=1
```

```python
#Top 10 Nodes of EigenVector Centrality
i=0

NIdEigenH = snap.TIntFltH()
snap.GetEigenVectorCentr(Graph, NIdEigenH)


NIdEigenH.SortByDat(False)
for item in NIdEigenH:
    print "node: %d ,EigenVectorCentr: %f" % (item, NIdEigenH[item])
    if(i>=9):
        break
    else:
        i+=1
```
