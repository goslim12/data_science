<li>로지스틱 회귀분석에 의한 분류의 정확도를 최종적으로 출력하시오.  </li>
</ul>

```python
from sklearn import linear_model
```

```python
sdf = df[['temp', 'RH',  'area']]
sdf
```

```python
sdf['area'] = np.log1p(sdf['area'])
```

```python
sdf.loc[:, 'fire'] = sdf['area'].apply(lambda x: 1 if x >= 0.418710 else 0)
```

```python
sdf
```

```python
attributes = sdf.iloc[:, 0:2]
area_data = attributes.values
area_data
```

```python
target_series = sdf.iloc[:, 2]
area_target = target_series.values
area_target
```

```python
regr = linear_model.LogisticRegression()
regr.fit(area_data, area_target)
```

```python
print 'Coefficients:', regr.coef_
print 'Intercept:', regr.intercept_
```

```python
predicted = []
for i in range(0, len(area_data)):
    predicted.append(regr.predict(area_data[i]))

sdf.loc[:,'predicted'] = predicted
sdf
```

- 로지스틱 회귀분석에 의한 분류 정확도

```python
correctly_predicted = 0
for i in range(0, len(area_data)):
    if predicted[i] == area_target[i]:
        correctly_predicted += 1
        
print float(correctly_predicted) / len(area_data)
```
